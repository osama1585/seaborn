<span style=color:green>Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.</span>

Ans-

***Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Here are five commonly used plots in Seaborn along with their uses:***

***Scatter plot:***

Use: Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents a single observation.

***Line plot:***

Use: Line plots are useful for visualizing the trend of a continuous variable over a specific range. They are often used to show the change in a variable over time.

***Bar plot:***

Use: Bar plots are effective for displaying and comparing the mean or aggregate values of different categories. They are commonly used for categorical data.

***Histogram:***

Use: Histograms are used to represent the distribution of a single variable. They provide a visual summary of the underlying frequency distribution.

***Box plot:***

Use: Box plots are used for visualizing the distribution of a continuous variable and highlighting the presence of outliers. They are particularly helpful in comparing the distributions of multiple groups.

<span style=color:green>Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.</span>

Ans-

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri_data = sns.load_dataset("fmri")

plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Dataset: Line Plot of Signal over Time for Different Events and Regions")

plt.show()
```

<span style=color:green>Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.</span>

Ans-

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_=sns.load_dataset('titanic')

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.boxplot(x="pclass",y="age" , data=titanic_)
plt.title("box plot of age by pclass")

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_)
plt.title('Box Plot of Fare by Pclass')


plt.show()
```

<span style=color:green>Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.</span>

Ans-

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_=sns.load_dataset('diamonds')

plt.figure(figsize=(10,6))
sns.histplot(data=diamonds_, x="price",hue="cut",multiple="stack", kde=True)

plt.xlabel("plot")
plt.ylabel("count")
plt.title("histogram of diamond prices")

plt.show()



```

<span style=color:green>Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.</span>

Ans-

```python
import seaborn as sns

iris=sns.load_dataset("iris")

sns.pairplot(iris,hue="species")

plt.show()
```
